```python
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'cyan', 'magenta']
print(colors[0])
```

```python
df = pd.read_csv("de_mirage_labeled_t_ct.csv")
```

```python
df = df.dropna(subset=['ct_role'])
```

```python
features = df[['t_opening', 't_firepower']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(features)
```

```python
plt.figure(figsize=(12, 8))
```

```python
markers = ['o', 's', '^', 'D', 'v', '*', 'P', 'X', 'h', '+']
```

```python
matplotlib inline
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
df['cluster'] = kmeans.fit_predict(X_scaled)
print(colors)
```

```python
plt.figure(figsize=(10, 6))
```

```python
roles = df['ct_role'].unique()
```

```python
markers = ['o', 's', '^', 'D', 'v', '*', 'P', 'X', 'h', '+']
```

```python
for i, role in enumerate(roles):
    subset = df[df['ct_role'] == role]
    plt.scatter(subset['ct_utility'], subset['ct_firepower'], 
                label=role, s=60, alpha=0.7, color=colors[i], marker=markers[i % len(markers)])
    for _, row in subset.iterrows():
        plt.text(
            row['ct_utility'],
            row['ct_firepower'],
            row['player_name'],
            fontsize=8,
            alpha=1
        )
```

```python
for i, role in enumerate(roles):
    subset = df[df['ct_role'] == role]
    plt.xlabel("CT Utility")
    plt.ylabel("CT Firepower")
    plt.title("K-Means Clustering by ct_utility & ct_firepower (colored by ct_role)")
    plt.scatter(subset['ct_utility'], subset['ct_firepower'], 
                label=role, s=60, alpha=0.7, color=colors[i], marker=markers[i % len(markers)])
    for _, row in subset.iterrows():
        plt.text(
            row['ct_utility'],
            row['ct_firepower'],
            row['player_name'],
            fontsize=8,
            alpha=0.6
        )
    plt.legend()
    plt.grid(True)
    plt.show()
```
