```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

# 
plt.scatter(df["age"], df["firepower"], color="blue", label="")

#    
for i, row in df.iterrows():
    plt.text(row["age"]+0.2, row["firepower"], str(row["age"]), fontsize=9)

#   ( )
coef = np.polyfit(df["age"], df["firepower"], 1)   #  (a, b)  ax + b
poly1d_fn = np.poly1d(coef)

#    min(age)  max(age)
x_vals = np.linspace(df["age"].min(), df["age"].max(), 100)
plt.plot(x_vals, poly1d_fn(x_vals), color="red", label="Trendline")

plt.xlabel("Age")
plt.ylabel("Firepower")
plt.title(" Firepower  Age")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```
